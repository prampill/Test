```python
def word_count(str):
    counts = dict()
    words = str.split()

    for word in words:
        if word in counts:
            counts[word] += 1
        else:
            counts[word] = 1

    return counts
```

```python
def word_counting(sentence):
    wor=[]
    for i in sentence.split(" "):
        wor.append(i.lower().strip('.'))

    cntwrd={}
    for j in wor:
        if j in cntwrd:
            cntwrd[j]=cntwrd[j]+1
        else:
            cntwrd[j]=1
    return cntwrd
```
